# Tag Analysis

This notebook analyzes the frequency of tags from the `tag_frequencies.json` file.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load tag frequencies
with open('../tags/tag_frequencies.json', 'r') as file:
    tag_freq = json.load(file)

# Convert to DataFrame
tag_df = pd.DataFrame(list(tag_freq.items()), columns=['Tag', 'Frequency'])

# Plot histogram to see distribution
plt.figure(figsize=(10, 5))
plt.hist(tag_df[tag_df['Frequency'] > 10]['Frequency'], bins=100, edgecolor='black', alpha=0.7)  # 50 bins for better granularity
plt.title('Distribution of Tag Frequencies')
plt.xlabel('Tag Frequency')
plt.ylabel('Count of Tags')
plt.yscale("log")  # Log scale to handle large differences
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import json
import pandas as pd

# Load tag frequencies
with open('../tags/tag_frequencies.json', 'r') as file:
    tag_freq = json.load(file)

# Convert to DataFrame
tag_df = pd.DataFrame(list(tag_freq.items()), columns=['Tag', 'Frequency'])

# Filter tags with frequency greater than 10
filtered_tag = tag_df[tag_df['Frequency'] > 100]['Tag']

# Save filtered tags to a text file
with open('../tags/filtered_tags.txt', 'w', encoding='utf-8') as out_file:
    for tag in filtered_tag:
        out_file.write(tag + '\n')
```
